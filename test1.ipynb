#   Show a plot of the response rate by the sales channel.




```python
# Calculate the count of responses by Sales Channel
total = data.groupby('Sales Channel')['Response'].count().reset_index(name='Total')

# Calculate the count of each response type by Sales Channel
response_count = data.groupby(['Sales Channel', 'Response']).size().reset_index(name='Count')

# Merge the total counts with the response counts
merged = pd.merge(response_count, total, on='Sales Channel')

# Calculate the percentage
merged['Percentage'] = (merged['Count'] / merged['Total']) * 100

# Create the plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Sales Channel', y='Percentage', hue='Response', data=merged)

# Set plot title and labels
plt.title('Percentage of Responses by Sales Channel')
plt.ylabel('Percentage (%)')
plt.xlabel('Sales Channel')

# Show the plot
plt.show()
```

#   Show a plot of the response rate by the total claim amount.[link text]


```python
#Andrés

max_income = data['Income'].max()

# Create bins starting from 0 up to the maximum income
bins = [0, max_income / 5, 2 * max_income / 5, 3 * max_income / 5, 4 * max_income / 5, max_income]

# Calculating the response rate by income with specified bins
response_rate_by_income = data.groupby(pd.cut(data['Income'], bins=bins))['Response'].apply(lambda x: (x == 'Yes').sum() / len(x) * 100)

# Resetting the index to use in Seaborn
response_rate_df = response_rate_by_income.reset_index()
response_rate_df.columns = ['Income Range', 'Response Rate']

# Creating the Seaborn bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Income Range', y='Response Rate', data=response_rate_df, palette='Blues_d')

# Setting plot title and labels
plt.title('Response Rate by Income')
plt.xlabel('Income Range')
plt.ylabel('Response Rate (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()

# Display the plot
plt.show()
```

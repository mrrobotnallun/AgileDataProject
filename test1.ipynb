
#   Show a plot of the total number of responses ✅


```python
response_count = data['Response'].value_counts()
print (response_count)

// Solution lab 2

#Show a plot of the total number of responses.

sns.histplot(x="Response",  data=data)
plt. title ('Count of Response')

#plt. xlabel ('Response')
#plt. ylabel ('Count')

plt.show()

// Matheus way

plt.figure(figsize=(6, 4))
response_count.plot(kind='bar', color='skyblue')
plt.title('Total Number of Responses')
plt.xlabel('Response')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
